```python
import yfinance as yf
import pandas as pd
    
```

```python
tickers = {
    "Indonesia (IDR)": "IDRUSD=X",
    "Malaysia (MYR)": "MYRUSD=X",
    "Thailand (THB)": "THBUSD=X",
    "Filipina (PHP)": "PHPUSD=X"
}

start_date = "2000-01-01"
end_date = "2024-12-31"
```

```python
data = {}

for country, ticker in tickers.items():
    df = yf.download(
        ticker,
        start=start_date,
        end=end_date,
        progress=False
    )

    # CEK APA DATA ADA
    if df.empty:
        print(country, "DATA KOSONG")
        continue

    # PASTIKAN ambil kolom yang benar
    series = df["Close"]

    # DROP NaN
    series = series.dropna()

    print(country, type(series), series.shape)

    data[country] = series
```

```python
exchange_df = pd.concat(data, axis=1)
exchange_df.head()
```

```python
exchange_df.to_csv("exchange-rate-project/data/exchange_rate.csv")
```
